```python
import matplotlib.pyplot as plt
import numpy as np

```

Basic Line Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

Styled Line Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

Bar Chart with NumPy and Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Salary', 'Expense', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

Scatter Plot with NumPy and Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y, color='purple', s=200)
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Histogram with NumPy and Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

Subplot Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

Save Figure Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color='green')
plt.title("Save Figure Example")

plt.savefig('my_plot.png')
plt.show()
```

Multiple Line Plot

```python
x = np.array([1, 2, 3, 4])

y1 = np.array([10, 15, 20, 25])
y2 = np.array([8, 12, 18, 22])

plt.plot(x, y1, label='Dataset A', color='red', marker='o')
plt.plot(x, y2, label='Dataset B', color='blue', marker='s', linestyle='--')

plt.title("Matplotlib with NumPy Arrays")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.legend()
plt.grid(False)

plt.show()
```

Pie Chart

```python
labels = np.array(['Food', 'Rent', 'Transport', 'Internet'])
sizes = np.array([40, 30, 20, 10])

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Expense Distribution")
plt.show()
```

Box Plot

```python
data = np.array([12, 15, 18, 20, 22, 24, 26, 30, 35])

plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

Area Chart

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 13, 20, 25])

plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.plot(x, y, color='blue')

plt.title("Area Chart Example")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

Stack Plot

```python
x = np.array([1, 2, 3, 4, 5])

y1 = np.array([3, 4, 5, 6, 7])
y2 = np.array([2, 3, 4, 5, 6])
y3 = np.array([1, 2, 1, 2, 3])

plt.stackplot(x, y1, y2, y3,
              labels=['Product A', 'Product B', 'Product C'])

plt.legend(loc='upper left')
plt.title("Stack Plot Example")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```
